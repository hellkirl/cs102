### Задание 1

1. Загрузите датасет `music_genre.csv`.
2. Возьмите 20 тысяч случайных строк из датасета (**все остальные задания выполняйте для этих 20 тысяч строк!!!**)
3. Проверьте, есть ли в датасете дублирующиеся строки. Если есть, удалите их.
4. Проверьте, есть ли а датасете строки, в которых значение в столбце `artist_name` – `empty_field`. Если есть, выведите эти строки и их количество. А затем удалите из датасета.

```python
import pandas as pd

"""Импорт датасета - шаг 1.1"""
dataframe = pd.read_csv("music_genre.csv")

"""Выборка случайных 20 тыс. строк - шаг 1.2"""
df = dataframe.sample(n=20000, random_state=2023)

"""Удаление дублирующийхся строк - шаг 2.3"""
df = df.drop_duplicates()

"""Удаление 'empty_field' строк - шаг 2.4"""
df = df[df.artist_name != "empty_field"]

df
```

Я импортировал csv-файл, случайным образом выбрал 20000 тысяч строк из датасета с помощью метода sample(), в аргументы которого для удобства добавил random_state, удалил дублирующиеся строки, а также удалил строки со значениями "empty_field".

### Задание 2

1. Выведите долю каждого жанра в датасете.
2. Покажите на горизонтальной столбиковой диаграмме распределение количества треков по жанрам. Выделите контрастным цветом жанр, в котором больше всего песен.
3. Найдите самый танцевальный жанр. У самого танцевального жанра среднее значение показателя danceability должно быть наивысшим.
4. Для каждого жанра определите, какая тональность в нем преобладает – количество треков какой тональность больше (минор или мажор).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

"""Доля каждого жанра - шаг 2.1"""
genre_percentage = df["music_genre"].value_counts() / df["music_genre"].value_counts().sum() * 100
pd.DataFrame(genre_percentage)

"""Столбиковая диаграмма - шаг 2.2"""
genres = df["music_genre"].value_counts().index
values = df["music_genre"].value_counts().values
df_genre_val = pd.DataFrame({"Music genre": genres, "Values": values})
cc = ['grey' if (x < max(values)) else 'red' for x in values]
plt.grid()
sns.barplot(y="Music genre", x="Values", data=df_genre_val, orient="h", palette=cc, )

"""Самый танцевальный жанр - шаг 2.3"""
danceability_percentage = df[["danceability", "music_genre"]]
danceable_genre = danceability_percentage.groupby("music_genre").apply(np.mean).sort_values(by=["danceability"]).reset_index().tail(1).reset_index(drop=True)

"""Тональность в жанре - шаг 2.4"""
genre_mode = df[["music_genre", "mode"]].groupby("music_genre").value_counts()
print(danceable_genre)
print(genre_mode)
```

Вывод:
Вычислив долю жанров в нашем датафрейме, я вывел столибковый график, в котором видно, что хип-хоп занимает перое место по количеству треков.

```python
import sqlite3

# создаем таблицу в БД
con = sqlite3.connect(":memory:")
cur = con.cursor()
cur.execute("CREATE TABLE music (artist, track, genre);")

# переносим в таблицу данные
cur.executemany("INSERT INTO music (artist, track, genre) "
                "VALUES (?, ?, ?);",
                df[["artist_name", "track_name", "music_genre"]].values.tolist())

# сохраняем изменения
con.commit()

# создаем и выполняем запрос
get_songs_per_artist = "SELECT artist, " \
                       "COUNT(artist) AS number_of_songs " \
                       "FROM music " \
                       "GROUP BY artist " \
                       "ORDER BY number_of_songs " \
                       "DESC;"
cur.execute(get_songs_per_artist)

result = cur.fetchall()
result[:20]
```
